## <a name="1">1. Importação</a>

```python
from matplotlib import pyplot as plt
import numpy as np
```

## <a name="2">2. Gráficos Simples</a>

```python
# Gráfico linear
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
# Gráfico linear X,Y
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
# definição do estilo da linha e dimensão eixos
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

## <a name="3">3. Gráficos usando keywords</a>

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
# c=color, s=marker_size
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

## <a name="4">4. Gráficos com categorias</a>

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
plt.figure(1, figsize=(9, 3))
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

## <a name="5">5. Estilos de linha</a>

```python
x=[1, 2, 3, 4]
y=[1, 4, 9, 16]
# usar keywords 
plt.plot(x, y, linewidth=2.0)
```

```python
# usar setter methods
line, = plt.plot(x, y, '-')
line.set_antialiased(False)
```

```python
x1 = np.array(x)
y1 = np.array(y)
x2 = x1 + 2
y2 = y1 - 5
lines = plt.plot(x1, y1, x2, y2)
plt.setp(lines, color='r', linewidth=2.0)
# or MATLAB style string value pairs
plt.setp(lines, 'color', 'r', 'linewidth', 2.0)
plt.show()
```

## <a name="6">6. Múltiplas figuras e eixos</a>

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.show()
```

```python
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## <a name="7">7. Trabalhar com texto</a>

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
```

```python
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts', fontsize=14, color='red')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max',xy=(2,1),xytext=(3,1.5), arrowprops=dict(facecolor='black',shrink=0.05),)

plt.ylim(-2, 2)
plt.show()
```

## <a name="8">8. Eixos logarítmicos e outros</a>

```python
from matplotlib.ticker import NullFormatter  # useful for `logit` scale

np.random.seed(19680801)
y=np.random.normal(loc=0.5, scale=0.4,size=1000)
y=y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure(1)
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
